# Reading Data

```python
g_r = np.genfromtxt('rdf_nvt.dat',skip_header=4)
eq_data = np.genfromtxt('equilibration.dat',skip_header=1)
pro_data = np.genfromtxt('production.dat',skip_header=1)
pot = np.genfromtxt('potential.dat',skip_header=4)
```

# Plots

```python
#plt.style.use('dark_background')
fig = plt.figure()
ax = fig.add_axes([0.15,0.15,0.80,0.80])
ax.plot(g_r[:,1],g_r[:,2],'--o',markersize = 5,linewidth = 2)
plt.xlabel("$r$",fontsize = 20)
plt.ylabel("$g(r)$",fontsize = 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid()
plt.savefig('g_r.jpg',dpi = 600)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.15,0.15,0.80,0.80])
ax.plot(eq_data[:,0],eq_data[:,3],alpha = 0.5,color = 'green')
plt.xlabel("$steps$",fontsize = 20)
plt.ylabel("$U$",fontsize = 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid()
plt.savefig('eq_pot.jpg',dpi = 600)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.15,0.15,0.80,0.80])
ax.plot(pro_data[:,0],pro_data[:,3],alpha = 0.5,color = 'crimson')
plt.xlabel("$steps$",fontsize = 20)
plt.ylabel("$U$",fontsize = 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid()
plt.savefig('pro_pot.jpg',dpi = 600)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.15,0.15,0.80,0.80])
ax.plot(eq_data[:,0],eq_data[:,1],alpha = 0.5,color = 'blue')
plt.xlabel("$steps$",fontsize = 20)
plt.ylabel("$T$",fontsize = 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid()
plt.savefig('temp.jpg',dpi = 600)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.15,0.15,0.80,0.80])
ax.plot(pot[:,1],pot[:,2],color = 'green',linewidth = 2)
plt.axhline(y=0,color = 'black',ls = '--')
plt.xlabel("$r$",fontsize = 20)
plt.ylabel("$U(r)$",fontsize = 20)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid()
plt.savefig('potential.jpg',dpi = 600)
```
